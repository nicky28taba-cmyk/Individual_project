# Data Science Planning Report: Predicting High-Volume Data Contributors
## Minecraft Server Player Analysis

**Research Question:** Which player characteristics predict whether a player will become a high-volume data contributor?

**Broad Question:** We would like to know which "kinds" of players are most likely to contribute a large amount of data so that we can target those players in our recruiting efforts.

---

## 1. DATA DESCRIPTION

### Dataset Overview
- **Players dataset:** 196 unique players with demographic information
- **Sessions dataset:** 1,535 individual play sessions
- Data collected automatically through server-side logging

```r
player_data |>
  select(played_hours, Age)|>
  summarise(mean_hours = round(mean(played_hours, na.rm = TRUE), 2),
        sd_hours = round(sd(played_hours, na.rm = TRUE), 2),
        mean_age = round(mean(Age, na.rm = TRUE), 2),
        sd_age = round(sd(Age, na.rm = TRUE), 2))

player_data |>
  count(experience)

player_data |>
  count(gender)

player_data |>
  count(subscribe)
```

```r
player_data |>
  ggplot(aes(x = played_hours)) +
  geom_histogram(bins = 30, color = "black", fill = "steelblue", alpha = 0.7) +
  labs(title = "Distribution of Total Hours Played", x = "Hours Played", y = "Number of Players")
```

```r
player_data |>
  ggplot(aes(x = experience, y = played_hours, fill = experience)) +
  geom_boxplot(alpha = 0.7) +
  labs(title = "Hours Played by Experience Level", x = "Experience Level", y = "Hours Played")
```

```r
threshold_75 <- quantile(player_data$played_hours, 0.75, na.rm = TRUE)

# Create binary indicator
player_data <- player_data %>%
  mutate(high_volume = if_else(played_hours >= threshold_75, TRUE, FALSE))

# Check results
player_data %>%
  count(high_volume)
```

```r
player_data |>
  group_by(experience) |>
  summarise(pct_high_volume = mean(high_volume) * 100) |>
  ggplot(aes(x = experience, y = pct_high_volume, fill = experience)) +
  geom_col() +
  labs(
    title = "Percentage of High-Volume Contributors by Experience",
    x = "Experience Level",
    y = "High-Volume %"
  ) +
theme(text = element_text(size = 15))
```

```r
player_data |>
  ggplot(aes(x = Age, y = played_hours, color = high_volume)) +
  geom_point(alpha = 0.6, size = 3) +
  labs(
    title = "Age vs Hours Played by High-Volume Status",
    x = "Age (years)",
    y = "Hours Played",
    color = "High-Volume"
  ) +
  scale_color_manual(values = c("TRUE" = "green", "FALSE" = "red"))
```


### Specific Research Question
**Can player demographic and behavioral characteristics (experience level, age, gender, and current playtime) predict whether a player will become a high-volume data contributor?**

### Why This Question Matters
The UBC Minecraft research server aims to collect behavioral data from gaming players. However, players don't contribute equally when taken a look at as a whole. Understanding which player types become high-volume contributors helps the research group with:

1. Recruiting players who are most likely to generate substantial data
2. Allocating server resources efficiently based on expected player engagement
3. Maximizing research data quality by focusing on engaged players
4. Plan for growth by predicting how new players with specific characteristics will contribute

### How We Define "High-Volume Contributor"
We define high-volume contributors as players in the **top 25% of total hours played** on the server. This binary classification allows us to:
- Easily identify and segment players
- Use logistic regression for prediction
- Distinguish between casual and serious contributors

### Variables Used
- `high_volume`: Binary indicator (TRUE if player is top 25%, FALSE otherwise)
- `experience`: Player skill level (Amateur, Beginner, Pro, Regular, Veteran)
- `Age`: Player age in years (demographic characteristic)
- `gender`: Gender identity (Male, Female, Other, Agender, Prefer not to say, Two-Spirited)
- `played_hours`: Total hours already invested (engagement metric)

### Expected Insights
By analyzing the data, we expect to discover:
- Whether professional/experienced players are more likely to become high-volume contributors
- If age is a significant predictor of contribution level
- Whether gender affects contribution patterns
- How current engagement (played_hours) relates to becoming a high-volume contributor

## Answer to Research Question

Based on the exploratory data analysis of the Minecraft server player data, it is clear that the answer to the specific question is in fact yes. Player demographic and behavioral characteristics do predict whether a player will become a high-volume data contributor. The visualizations reveal that experience level is the strongest predictor of high-volume contribution. Professional players have the highest percentage of high-volume contributors at approximately 35%, followed by beginner players at 33% and Regular players at 27%, while amateur and veteran players show lower rates at 25% and 18% respectively. Age also emerges as an important predictor, with high-volume contributors (shown in green) predominantly concentrated in the 15-30 year age range, while low-volume contributors (shown in red) are dispersed across all age groups. The distribution of hours played is heavily right-skewed, indicating that while most players are casual contributors, a meaningful segment (approximately 25% or 144 out of 196 players) become high-volume contributors with substantially more engagement. The box plot further confirms that professional, beginner, and regular players maintain higher median hours played compared to amateur and veteran players. These patterns suggest that targeting recruitment efforts toward younger, professionally-skilled players aged 15-30 would likely yield the highest data contribution rates for the research group, making this demographic profile valuable for optimizing server resource allocation and recruitment strategy.
